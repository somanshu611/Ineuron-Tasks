> 10. from a cabin column seperate text and numeric value .

```python
# Approach 1 : 
import re

replace = re.compile('([a-zA-Z]+)')

df2 = df
df2['str_cabin_list'] = df2['Cabin'].str.extract(replace)
df2['num_cabin_list'] = df2['Cabin'].str.replace(replace,"")
```

```python
# OR, you don't need to import re package

df2['str_cabin'] = df2['Cabin'].str.extract('([A-Za-z]+)')
df2['num_cabin'] = df2['Cabin'].str.replace('([A-Za-z]+)','')
```

```python
df2
```

```python
# Approach 2 : 
# The second part of the regular expression ( (^[a-zA-z]+) ), is checking for NOT STRING, only numbers
df3 = pd.DataFrame(df['Cabin'].str.split('([a-zA-Z]+)(^[a-zA-Z]+)'))
```
